```python
import pandas as pd
import matplotlib.pyplot as plt

columns =['head_neck_inci', 'colon_inci', 'rectum_inci', 'oesophageal_inci', 'stomach_inci', 'proximal_colon_inci', 'distal_colon_inci', 'gallbladder_inci', 'pancreatic_inci', 'larynk_inci', 'lung_inci', 'breast_inci', 'uterine_inci', 'ovarian_inci', 'prostate_inci', 'kidney_inci', 'bladder_inci', 'malignant_melanoma_inci', 'cervix_inci', 'testis_inci', 'brain_inci', 'tyroid_inci', 'hodgkin_lymphoma_inci', 'non_hodgkin_lymphoma_inci', 'multiple_myeloma_inci', 'leukemia_inci']

counts = []

# Count the occurrences of 'yes' in each column
for column in columns:
    counts.append(cancer[column].value_counts()['yes'])

# Sort the counts in descending order
sorted_counts = sorted(counts, reverse=True)
sorted_columns = [x for _, x in sorted(zip(counts, columns), reverse=True)]

# Create a bar plot for the sorted counts of 'yes' values
plt.figure(figsize=(15, 6))
plt.bar(sorted_columns, sorted_counts)
plt.xticks(rotation=90)
plt.xlabel('Cancer Type', labelpad=10,color='green')
plt.ylabel('Count')
plt.title('Counts of "Yes" Incidence_cancer_filter_end (Highest to Lowest)')

# Add value labels on top of each bar
for i, v in enumerate(sorted_counts):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

```python
data['date_of_death'] = pd.to_datetime(data['date_of_death'], errors='coerce')
data['date_end_study'] = pd.to_datetime(data['date_end_study'], errors='coerce')
data['datecancer_22'] = pd.to_datetime(data['datecancer_22'], errors='coerce')

import numpy as np

data['duration_cancer22'] = np.where(
    data['date_of_death'].notnull(), 
    round(abs((data['date_of_death'] - data['datecancer_22']).dt.days / 365.25), 1),
    np.where(data['date_end_study'].notnull(), 
             round(abs((data['date_end_study'] - data['datecancer_22']).dt.days / 365.25), 1), 
             np.nan)
)
```
